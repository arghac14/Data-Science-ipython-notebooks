```
from matplotlib import pyplot as plt
from matplotlib import style
```

## Line Graph

```
x=[1,2,5,7,9]
y=[1,4,6,7,8]

plt.plot(x,y)

plt.title('x vs y')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```

plt.style.use('ggplot')    #'dark_background','presentation'

x1=[1,2,3,4,5,6]
y1=[1,4,5,7,9,11]
x2=[2,3,4,5,6,7]
y2=[3,5,6,7,10,12]
   
plt.plot(x1,y1,'g',label='line1',linewidth=4)
plt.plot(x2,y2,'g',label='line1',linewidth=4)
    

plt.title('x vs y')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.grid(True,color='k')
plt.show()
```

## Bar graph

```
x=[67,28,79,90]
y=[37,13,39,89]
x1=[28,19,20,13]
y1=[90,89,78,56]

plt.bar(x,y,label='ex 1',color='r')
#plt.barh(x1,y1,label='ex 1',color='g')  #for horizontal bars


plt.title('x vs y',color='b')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.grid(True,color='k')
plt.show()
```

## Histogram (?)

```
x=[12,40,60,85]
y=[14,45,82,178]

plt.hist(x,y,histtype='bar',rwidth=0.2)

plt.title('x vs y',color='b')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.grid(True,color='k')
plt.show()
```

## Scatter plot

```
x=[1,2,3,4,5,6]
y=[10,31,4,13,2,93]

plt.scatter(x,y,label='',color='b')

plt.title('x vs y',color='g')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.grid(True,color='k')
plt.show()
```

# Area/stack plot (?)

```
days=[1,2,3]
sleeping=[10,9,11]
eating=[1,0.2,0.4]
working=[8,9,7]
playing=[1,1.4,1.6]

plt.plot(days,sleeping,color='m',label='sleeping',linewidth=4)
plt.plot(days,eating,color='g',label='eating',linewidth=4)
plt.plot(days,working,color='r',label='working',linewidth=4)
plt.plot(days,playing,color='b',label='sleeping',linewidth=4)

plt.stackplot(days,sleeping,eating,working,playing)
colors=['m','g','r','b']

plt.title('x vs y',color='g')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.grid(True,color='k')
plt.show()
```

## Pie chart

```
slices=[1,2,3]
activities=['sleeping','coding','playing']
colors=['c','m','r']

plt.pie(slices,labels=activities,colors=colors,startangle=90,shadow=True,explode=(0,0.1,0),autopct='%1.1f%%')
```
