## Groupby

```python
myData={
    'franchise':['Marvel','DC','Marvel','Marvel','DC','DC','Marvel','DC'],
    'team':['Avengers','Justice League','Avengers','Avengers','Justice League','Justice League','Avengers','Justice League'],
    'hero':['Ironman','Batman','Captain America','Thor','Superman','Wonder Woman','Spider-man','Aquaman'],
    'power-level':[280,260,220,870,1230,760,180,240]
  }
df=pd.DataFrame(myData)
df
```

```python
df1=df.groupby('team')
df1
```

# Concate

```python
kolkata={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "wind-speed":[12,10,9,14],
      "temp":[14,16,15,10]
}
kol=pd.DataFrame(kolkata)
Mumbai={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "wind-speed":[12,10,9,14],
      "temp":[14,16,15,10]
}
mum=pd.DataFrame(Mumbai)

df=pd.concat([kol,mum])
df
```

```python
df=pd.concat([kol,mum],ignore_index=True)
df
```

```python
df=pd.concat([kol,mum],keys=['Kolkata','Mumbai'])
df
```

```python
df.loc['Mumbai']
```

```python
df=pd.concat([kol,mum],axis=1)
df
```

# Merge

```python
temp1={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "temp":[14,16,15,10]
}
temp=pd.DataFrame(temp1)
ws={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "wind-speed":[12,10,9,14],
}
wind_speed=pd.DataFrame(ws)
```

```python
#Inner join (In this method you will get the intersetion of two dataframes with merged column.It is the default merge method.)

df=pd.merge(temp,wind_speed,on=['date','event'])
df
```

```python
#Outer join (This is just like union of two dataframe.The value which dont exist will contain NaN.)
temp1={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','hot','sunny'],
      "temp":[14,16,15,10]
}
temp=pd.DataFrame(temp1)
ws={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "wind-speed":[12,10,9,14],
}
wind_speed=pd.DataFrame(ws)

df=pd.merge(temp,wind_speed,on=['date','event'],how='outer')
df
```
