```python
import pandas as pd
```

# Handling missing datas:

```python
df=pd.read_csv('D:\Courses\Data-analysis-with-pandas-master\datasets\season.csv')
df
```

## fillna method

```python
df1=df.fillna(0)   # 0 is the fill value
df1
```

```python
df1=df.fillna(method='ffill')     # forward fill method (fills the value of NaN with the previous value of the same column)
df1
```

```python
df1=df.fillna(method='bfill')       # backward fill method (fills the value of NaN with the next value of the same column)
df1
```

### Interpolation method

```python
df2=df.interpolate()   #linear interpolation (Other methods: quadritic,polynomial,time)
df2

#Note:
#-Only temp column gets affected because it contains values.
#-Interpolation only applies to columns which contains values not strings.
```

## dropna method

```python
df3=df.dropna()   #will drop all the rows which contains atleast one NaN.
df3
```

```python
df3=df.dropna(how='all')  #Droping all the rows which contains all the values NaN(except: index) 
df3
```

```python
df3=df.dropna(thresh=4)   #Minimum 4 valid values in all coloumns
df3
```

## replace method

```python
import numpy as np
df4=df.replace(["sunny"],np.NaN)
df4
```

```python
df4=df.replace({
    'day':['hot'],
    'wind-speed':['45','41']
},np.NaN)
df4
```

```python
df4=df.replace(['sunny','hot','rainy'],['warm','overcast','cloudy'])
df4
```
